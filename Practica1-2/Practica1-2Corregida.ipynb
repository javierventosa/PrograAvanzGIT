```python
#Ejercicio 1

for i in range(1, 6):
    if i % 2 == 0:
        continue  # Omitir números pares
    for j in range(1, 11):
        product = i * j
        if product > 20:
            break  # Detener si el producto supera 20
        print(f"{i} x {j} = {product}")
        pass  # Uso de pass sin afectar la lógica

#Ejercicio 2

def process_text(text="Este es un texto de ejemplo.", words_to_replace=None):
    """
    Procesa un texto dado, reemplazando palabras específicas con asteriscos.

    Parámetros:
    text (str): El texto a procesar. Por defecto es "Este es un texto de ejemplo.".
    words_to_replace (list): Una lista de palabras a reemplazar. Por defecto es ["texto", "ejemplo"].

    Devuelve:
    tuple: Una tupla que contiene el texto procesado y la cantidad de palabras reemplazadas.
    """
    if words_to_replace is None:
        words_to_replace = ["texto", "ejemplo"]

    # Convertir la cadena a minúsculas
    text = text.lower()

    # Eliminar los espacios en blanco iniciales y finales
    text = text.strip()

    # Reemplazar las palabras especificadas con asteriscos
    replaced_count = 0
    for word in words_to_replace:
        if word in text:
            text = text.replace(word, '*' * len(word))
            replaced_count += 1

    return text, replaced_count

def main():
    """
    Función principal que procesa un texto de ejemplo y muestra el resultado.

    Llama a la función process_text para procesar el texto y luego imprime el texto procesado
    y la cantidad de palabras reemplazadas.
    """
    processed_text, count = process_text()
    print(f"Texto procesado: {processed_text}")
    print(f"Cantidad de palabras reemplazadas: {count}")

if __name__ == "__main__":
    main()

#Ejercicio 3

def is_prime(n):
    """
    Comprueba si un número n es primo.

    Args:
    n (int): El número a comprobar.

    Returns:
    bool: True si n es primo, False en caso contrario.
    """
    if n <= 1:
        return False  # Porque los números primos son mayores a 1
    for i in range(2, int(n**0.5) + 1):  # Se busca un divisor hasta la raíz cuadrada de n
        if n % i == 0:
            return False  # Si n es divisible por algún número, no es primo
    return True  # Si no se encontró ningún divisor, n es primo

def prime_list(limit):
    """
    Genera una lista de números primos hasta un límite.

    Args:
    limit (int): El límite hasta el cual generar números primos.

    Returns:
    list: Lista de números primos hasta el límite.
    """
    primes = []
    for num in range(2, limit + 1):
        if is_prime(num):
            primes.append(num)
    return primes

def check_palindrome(primes):
    """
    Busca palíndromos en una lista de números primos.

    Args:
    primes (list): Lista de números primos.

    Returns:
    list: Lista de números primos que son palíndromos.
    """
    palindromes = []
    for prime in primes:
        if str(prime) == str(prime)[::-1]:  # Comprobar si el número es un palíndromo
            palindromes.append(prime)
    return palindromes

def main():
    """
    Función principal que genera números primos y busca palíndromos entre ellos.
    """
    limit = 100  # Puedes cambiar el límite según sea necesario
    primes = prime_list(limit)
    palindromes = check_palindrome(primes)
    print(f"Números primos hasta {limit}: {primes}")
    print(f"Números primos que son palíndromos: {palindromes}")

if __name__ == "__main__":
    main()def is_prime(n):
    """
    Comprueba si un número n es primo.

    Args:
    n (int): El número a comprobar.

    Returns:
    bool: True si n es primo, False en caso contrario.
    """
    if n <= 1:
        return False  # Porque los números primos son mayores a 1
    for i in range(2, int(n**0.5) + 1):  # Se busca un divisor hasta la raíz cuadrada de n
        if n % i == 0:
            return False  # Si n es divisible por algún número, no es primo
    return True  # Si no se encontró ningún divisor, n es primo

def prime_list(limit):
    """
    Genera una lista de números primos hasta un límite.

    Args:
    limit (int): El límite hasta el cual generar números primos.

    Returns:
    list: Lista de números primos hasta el límite.
    """
    primes = []
    for num in range(2, limit + 1):
        if is_prime(num):
            primes.append(num)
    return primes

def check_palindrome(primes):
    """
    Busca palíndromos en una lista de números primos.

    Args:
    primes (list): Lista de números primos.

    Returns:
    list: Lista de números primos que son palíndromos.
    """
    palindromes = []
    for prime in primes:
        if str(prime) == str(prime)[::-1]:  # Comprobar si el número es un palíndromo
            palindromes.append(prime)
    return palindromes

def main():
    """
    Función principal que genera números primos y busca palíndromos entre ellos.
    """
    limit = 100  # Puedes cambiar el límite según sea necesario
    primes = prime_list(limit)
    palindromes = check_palindrome(primes)
    print(f"Números primos hasta {limit}: {primes}")
    print(f"Números primos que son palíndromos: {palindromes}")

if __name__ == "__main__":
    main()

#Ejercicio 4

def sum_all(numbers):
    """
    Calcula la suma de todos los números en la lista.

    Args:
    numbers (list): Lista de números enteros.

    Returns:
    int: Suma de todos los números.
    """
    return sum(numbers)

def sum_positive_negative(numbers):
    """
    Calcula la suma de los números positivos y negativos por separado.

    Args:
    numbers (list): Lista de números enteros.

    Returns:
    tuple: Suma de los números positivos y suma de los números negativos.
    """
    positive_sum = sum(n for n in numbers if n > 0)
    negative_sum = sum(n for n in numbers if n < 0)
    return positive_sum, negative_sum

def find_max_min(numbers):
    """
    Encuentra el número máximo y mínimo de la lista.

    Args:
    numbers (list): Lista de números enteros.

    Returns:
    tuple: Número máximo y número mínimo.
    """
    return max(numbers), min(numbers)

def check_unique(numbers):
    """
    Comprueba si todos los números de la lista son únicos.

    Args:
    numbers (list): Lista de números enteros.

    Returns:
    bool: True si todos los números son únicos, False en caso contrario.
    """
    return len(numbers) == len(set(numbers))

def compare_positive_negative(numbers):
    """
    Devuelve un mensaje que indica si la lista contiene más números positivos o negativos.

    Args:
    numbers (list): Lista de números enteros.

    Returns:
    str: Mensaje indicando si hay más números positivos o negativos.
    """
    positive_count = sum(1 for n in numbers if n > 0)
    negative_count = sum(1 for n in numbers if n < 0)
    if positive_count > negative_count:
        return "La lista contiene más números positivos."
    elif negative_count > positive_count:
        return "La lista contiene más números negativos."
    else:
        return "La lista contiene igual cantidad de números positivos y negativos."

def main():
    """
    Función principal que implementa el menú y llama a las funciones correspondientes.
    """
    numbers = list(map(int, input("Introduce una lista de números enteros separados por espacios: ").split()))

    while True:
        print("\nMenú:")
        print("1. Calcula la suma de todos los números.")
        print("2. Calcula la suma de los números positivos y la suma de los números negativos por separado.")
        print("3. Encuentra el número máximo y el mínimo de la lista.")
        print("4. Comprueba si todos los números de la lista son únicos.")
        print("5. Indica si la lista contiene más números positivos o negativos.")
        print("6. Salir.")

        choice = input("Elige una opción: ")

        if choice == '1':
            print(f"Suma de todos los números: {sum_all(numbers)}")
        elif choice == '2':
            positive_sum, negative_sum = sum_positive_negative(numbers)
            print(f"Suma de los números positivos: {positive_sum}")
            print(f"Suma de los números negativos: {negative_sum}")
        elif choice == '3':
            max_num, min_num = find_max_min(numbers)
            print(f"Número máximo: {max_num}")
            print(f"Número mínimo: {min_num}")
        elif choice == '4':
            if check_unique(numbers):
                print("Todos los números de la lista son únicos.")
            else:
                print("Hay números repetidos en la lista.")
        elif choice == '5':
            print(compare_positive_negative(numbers))
        elif choice == '6':
            print("Saliendo del programa.")
            break
        else:
            print("Opción no válida. Por favor, elige una opción del 1 al 6.")

if __name__ == "__main__":
    main()

#Ejercicio 5

# Lista global de tareas
tasks = []

def add_task(task):
    """
    Agrega una nueva tarea a la lista global de tareas.

    Args:
    task (str): La tarea a agregar.
    """
    tasks.append(task)
    print(f"Tarea '{task}' agregada.")

def remove_task(task):
    """
    Elimina una tarea de la lista global de tareas, si existe.

    Args:
    task (str): La tarea a eliminar.
    """
    if task in tasks:
        tasks.remove(task)
        print(f"Tarea '{task}' eliminada.")
    else:
        print(f"Tarea '{task}' no encontrada.")

def list_tasks():
    """
    Enumera todas las tareas almacenadas en la lista global de tareas.
    """
    if tasks:
        print("Lista de tareas:")
        for idx, task in enumerate(tasks, 1):
            print(f"{idx}. {task}")
    else:
        print("No hay tareas en la lista.")

def task_manager():
    """
    Administra las operaciones de agregar, modificar y enumerar tareas.
    """
    def modify_task(new_task):
        """
        Modifica la primera tarea en la lista de tareas.

        Args:
        new_task (str): La nueva tarea que reemplazará la primera tarea.
        """
        if tasks:
            old_task = tasks[0]
            tasks[0] = new_task
            print(f"Tarea '{old_task}' modificada a '{new_task}'.")
        else:
            print("No hay tareas para modificar.")

    while True:
        # Mostrar el menú de opciones
        print("\nMenú de Tareas:")
        print("1. Agregar tarea")
        print("2. Eliminar tarea")
        print("3. Listar tareas")
        print("4. Modificar la primera tarea")
        print("5. Salir")

        choice = input("Elige una opción: ")

        if choice == '1':
            task = input("Introduce la tarea a agregar: ")
            add_task(task)
        elif choice == '2':
            task = input("Introduce la tarea a eliminar: ")
            remove_task(task)
        elif choice == '3':
            list_tasks()
        elif choice == '4':
            new_task = input("Introduce la nueva tarea para modificar la primera tarea: ")
            modify_task(new_task)
        elif choice == '5':
            print("Saliendo del programa.")
            break
        else:
            print("Opción no válida. Por favor, elige una opción del 1 al 5.")

def main():
    """
    Función principal que llama a task_manager para manejar las tareas.
    """
    task_manager()

if __name__ == "__main__":
    main()

#Ejercicio 6

# Lista global de cadenas
strings = []

def print_strings_starting_with(letter):
    """
    Imprime cada cadena que comience con una letra pedida al usuario.

    Args:
    letter (str): La letra con la que deben comenzar las cadenas.
    """
    for string in strings:
        if string.startswith(letter):
            print(string)

def count_strings_containing(substring):
    """
    Cuenta cuántas cadenas contienen una subcadena pedida al usuario.

    Args:
    substring (str): La subcadena a buscar en las cadenas.

    Returns:
    int: La cantidad de cadenas que contienen la subcadena.
    """
    count = sum(1 for string in strings if substring in string)
    return count

def find_longest_and_shortest():
    """
    Encuentra la cadena más larga y la más corta de la lista.

    Returns:
    tuple: La cadena más larga y la más corta.
    """
    if not strings:
        return None, None  # Manejo de error si la lista está vacía

    longest = max(strings, key=len)
    shortest = min(strings, key=len)
    return longest, shortest

def main():
    """
    Función principal que implementa el menú y llama a las funciones correspondientes.
    """
    while True:
        print("\nMenú de Cadenas:")
        print("1. Imprimir cadenas que comienzan con una letra específica")
        print("2. Contar cadenas que contienen una subcadena específica")
        print("3. Encontrar la cadena más larga y la más corta")
        print("4. Salir")

        choice = input("Elige una opción: ")

        if choice == '1':
            letter = input("Introduce la letra: ")
            print_strings_starting_with(letter)
        elif choice == '2':
            substring = input("Introduce la subcadena: ")
            count = count_strings_containing(substring)
            print(f"Cantidad de cadenas que contienen '{substring}': {count}")
        elif choice == '3':
            longest, shortest = find_longest_and_shortest()
            if longest is None and shortest is None:
                print("La lista de cadenas está vacía.")
            else:
                print(f"La cadena más larga es: '{longest}'")
                print(f"La cadena más corta es: '{shortest}'")
        elif choice == '4':
            print("Saliendo del programa.")
            break
        else:
            print("Opción no válida. Por favor, elige una opción del 1 al 4.")

if __name__ == "__main__":
    main()
```
